###Generating the job description using gpt 2



```python
#importing necessary classes from huggingface transformers library
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch # pytorch library used for tensor computations
```

```python
# Load pre-trained model and tokenizer (GPT-2 in this case)
model_name = "gpt2-large"  # You can also try "gpt2-medium", "gpt2-large", etc.
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
def generate_job_description(Job_title, Job_summary, Responsibilities, Education, Experience, Skills):
  """Function to generate job description."""

  # Tokenize the prompt
  prompt= f"""
  Job Title: {Job_title}
  Job Summary: {Job_summary}
  Responsibilities: {Responsibilities}
  Education: {Education}
  Experience: {Experience}
  Skills: {Skills}

  """

  input_ids = tokenizer.encode(prompt, return_tensors="pt")

  # Set the attention mask and the pad token id
  attention_mask = torch.ones_like(input_ids)
  pad_token_id = 0

  # Generating the job description using the model
  output = model.generate(input_ids, attention_mask=attention_mask, pad_token_id=pad_token_id, max_length=495, num_return_sequences=1, no_repeat_ngram_size=2, top_k=900)

  # Decoding the generated tokens
  generated_jd_text = tokenizer.decode(output[0], skip_special_tokens=True)

  return generated_jd_text
```

```python
# Getting the input from the user
job_title = input("Enter the job title: ")
job_summary = input("Enter an overview of the job role: ")
responsibilities = input("The responsibilities of the candidate in the job role: ")
education = input("Enter the Educational requirements: ")
experience = input("Preferred years of experience: ")
skills = input("Skills required for the job: ")
```

```python
# Generating the job description
generated_job_description = generate_job_description(job_title, job_summary, responsibilities, education, experience, skills)

# Print the generated job description
print(generated_job_description)
```

### Comments: After changing the parameters max_length(no. of words generated), num_return_sequences(no. of job_descriptions), no_repeat_ngram_size(repeatability of words), top_k(exploring while generating words). the model was generate the above results. Given more time the model can be modified and hyperparameters can be changed to get more accuracy
